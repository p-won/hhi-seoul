# 카드·배달 데이터 전처리 및 행정동별_최종분석_v3 조인

```python
import pandas as pd

# ── 경로 설정 ──────────────────────────────────────────────
CARD_DELIVERY_PATH = "data/최종결과.csv"            # R 코드로 만든 파일 경로
BASE_PATH          = "data/행정동별_최종분석_v3.csv"
OUTPUT_PATH        = "data/행정동별_최종분석_v4.csv"
```

## 1. R 결과 파일 로드 & 행정동 개수 확인

```python
cd = pd.read_csv(CARD_DELIVERY_PATH, encoding='utf-8-sig')
print("컬럼:", cd.columns.tolist())
print("shape:", cd.shape)
cd.head()
```

```python
# 기준년월별 행정동 개수
print("기준년월 종류:", sorted(cd['기준년월'].unique()))
print("\n행정동 개수 (전체 unique):", cd['행정동명'].nunique())
print("\n행정동 목록:")
print(sorted(cd['행정동명'].unique()))
```

## 2. 기준년월 제거 → 행정동별 중앙값 집계

```python
target_cols = ['배달사용금액', '배달사용건수', '카드사용금액', '카드사용건수']

cd_agg = (
    cd.groupby('행정동명')[target_cols]
    .median()
    .reset_index()
)

print("집계 후 행정동 수:", len(cd_agg))
cd_agg.head(10)
```

## 3. 행정동명 매칭 확인 (v3 ↔ R 결과)

```python
base = pd.read_csv(BASE_PATH, encoding='utf-8-sig')
print("v3 shape:", base.shape)

base_dongs = set(base['행정동명'])
cd_dongs   = set(cd_agg['행정동명'])

print(f"\nv3 행정동 수  : {len(base_dongs)}")
print(f"R결과 행정동 수: {len(cd_dongs)}")
```

```python
# v3에는 있는데 R결과에 없는 행정동
only_in_base = sorted(base_dongs - cd_dongs)
print(f"v3에만 있는 행정동 ({len(only_in_base)}개):")
print(only_in_base)

# R결과에는 있는데 v3에 없는 행정동
only_in_cd = sorted(cd_dongs - base_dongs)
print(f"\nR결과에만 있는 행정동 ({len(only_in_cd)}개):")
print(only_in_cd)

# 공통
common = base_dongs & cd_dongs
print(f"\n공통 행정동 수: {len(common)}")
```

## 4. 조인 (left join 기준: v3)

```python
final = base.merge(cd_agg, on='행정동명', how='left')

print("조인 후 shape:", final.shape)

# 조인 후 누락 확인
missing_after_join = final[final['배달사용금액'].isna()]['행정동명'].tolist()
if missing_after_join:
    print(f"\n조인 후 카드·배달 값 없는 행정동 ({len(missing_after_join)}개):")
    print(missing_after_join)
else:
    print("\n모든 행정동 조인 완료 — 누락 없음")
```

## 6. v4 저장

```python
final.to_csv(OUTPUT_PATH, index=False, encoding='utf-8-sig')
print(f"저장 완료: {OUTPUT_PATH}  ({len(final)}행 × {len(final.columns)}열)")
```
